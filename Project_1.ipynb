```python
import turicreate
import matplotlib.pyplot as plt
```

```python
sf = turicreate.SFrame('home_data.sframe/')
```

```python
sf.head()
```

```python
turicreate.show(sf[1:5000]['sqft_living'],sf[1:5000]['price'])
```

```python
plt.scatter(x=sf['sqft_living'],y=sf['price'])
plt.show
```

```python
train_data, test_data = sf.random_split(.8,seed=0)
```

```python
sqft_model = turicreate.linear_regression.create(train_data,target='price',features=['sqft_living'])
```

```python
print(test_data['price'].mean())
```

```python
print(sqft_model.evaluate(test_data))
```

```python
plt.plot(test_data['sqft_living'],test_data['price'],'.',
         test_data['sqft_living'],sqft_model.predict(test_data),'-')
```

```python
features = ['bedrooms','bathrooms','sqft_living','sqft_lot','floors','zipcode']
```

```python
sf[features].show()
```

```python
features_model = turicreate.linear_regression.create(train_data,target='price',features=features)
```

```python
print(sqft_model.evaluate(test_data))
print(features_model.evaluate(test_data))
```
